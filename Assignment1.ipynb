```python
sns.set(style=('whitegrid'))
year_columns=['2022 Population',
 '2020 Population',
 '2015 Population',
 '2010 Population',
 '2000 Population',
 '1990 Population',
 '1980 Population',
 '1970 Population']
```

```python
pop_in_year=df[year_columns].sum().sort_index()
pop_in_year
```

```python
plt.figure(figsize=(12, 6))
plt.bar(pop_in_year.index, pop_in_year.values, color='skyblue')
plt.title('Population over year')
plt.ylabel('Population (in millions) per year')

plt.tick_params(axis='x', rotation=45)
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
top10_pop = df.sort_values(by='2022 Population', ascending=False).head(10)
top10_pop
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_pop, x='2022 Population', y='Country/Territory', palette='magma')
plt.title("Top 10 Most Populous Countries (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Population (in Millions)", fontsize=12, fontweight='bold')
plt.ylabel("Country", fontsize=12, fontweight='bold')
plt.tick_params(bottom=True)
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
continent_pop =df.groupby('Continent')['2022 Population'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=continent_pop.values, y=continent_pop.index, palette='magma')
plt.title(" Population of continents by 2022", fontsize=14, fontweight='bold')
plt.xlabel("Population (in Millions)", fontsize=12, fontweight='bold')
plt.ylabel("Continent", fontsize=12, fontweight='bold')
plt.tick_params(bottom=True)
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
top10_density_pop = df.sort_values(by='Density (per km²)', ascending=False).head(10)
top10_density_pop
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_density_pop, x='Density (per km²)', y='Country/Territory', palette='magma')
plt.title(" Top 10 density wise populated continent 2022", fontsize=14, fontweight='bold')
plt.xlabel("Population (in Millions)", fontsize=12, fontweight='bold')
plt.ylabel("Country", fontsize=12, fontweight='bold')
plt.tick_params(bottom=True)
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
top10_area_pop = df.sort_values(by='Area (km²)', ascending=False).head(10)
top10_area_pop
```
